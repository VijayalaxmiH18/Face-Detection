```
import cv2
```

```
#Load the cascade 
face_cascade = cv2.CascadeClassifier('/content/drive/My Drive/Colab Notebooks/facedetection-master/haarcascade_frontalface_default.xml')
```

```
#Read the input image 
img = cv2.imread('/content/drive/My Drive/Colab Notebooks/testimg.jpg')
```

```
#Convert RGB image into grauscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```
#Detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
```

```
#Draw rectangle around the faces 
for (x,y,w,h) in faces:
  cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
```
